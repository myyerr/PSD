# Business Understanding – Data Iris

## Latar Belakang
Identifikasi spesies bunga *Iris* secara manual membutuhkan keterampilan botani, sehingga bisa memakan waktu dan berpotensi terjadi kesalahan.  
Dengan memanfaatkan analisis data dan *machine learning*, identifikasi dapat dilakukan secara otomatis dan konsisten.  

Dataset *Iris* dalam proyek ini terhubung melalui beberapa sumber:  
- **CSV lokal**
- **phpMyAdmin (MySQL/MariaDB)**
- **PostgreSQL**

## Tujuan Proyek
- Mengembangkan model klasifikasi yang mampu mengenali spesies *Iris* (Setosa, Versicolor, Virginica) secara akurat.  
- Menganalisis karakteristik setiap parameter (Sepal Length, Sepal Width, Petal Length, Petal Width) untuk menemukan pola yang relevan dengan klasifikasi.  
- Menyediakan sistem yang fleksibel dengan data yang dapat diakses dari berbagai sumber (database maupun file lokal).  

## Manfaat yang Diharapkan
1. **Efisiensi Identifikasi** mempercepat klasifikasi spesies bunga tanpa perlu ahli botani.  
2. **Integrasi Data** memastikan data dapat diakses dari berbagai platform (CSV, phpMyAdmin, PostgreSQL).  
3. **Pengambilan Keputusan** hasil analisis dapat divisualisasikan (misalnya dengan Power BI) untuk mendukung keputusan ilmiah atau pembelajaran.  
4. **Studi Kasus Pendidikan** menjadi contoh standar penerapan CRISP-DM untuk proyek sains data berbasis klasifikasi.  

## Objektif
- **Objektif bisnis:** Membangun sistem klasifikasi berbasis data yang dapat membantu proses identifikasi spesies bunga *Iris* dengan akurat dan efisien.  
- **Objektif data:** Mengintegrasikan dataset dari berbagai sumber, menganalisis nilai min, max, dan average setiap parameter, lalu membangun model machine learning untuk klasifikasi.  


# Data Understanding – Data Iris

## Gambaran Dataset
Dataset *Iris* berisi data karakteristik bunga *Iris* dengan total 150 baris dan 5 kolom utama:
| Kolom             | Tipe Data   | Deskripsi                                                    |
| ----------------- | ----------- | ------------------------------------------------------------ |
| **Id**            | SERIAL      | Nomor urut unik untuk setiap baris data (*primary key*).     |
| **SepalLengthCm** | FLOAT       | Panjang sepal (kelopak bunga) dalam satuan cm.               |
| **SepalWidthCm**  | FLOAT       | Lebar sepal (kelopak bunga) dalam satuan cm.                 |
| **PetalLengthCm** | FLOAT       | Panjang petal (mahkota bunga) dalam satuan cm.               |
| **PetalWidthCm**  | FLOAT       | Lebar petal (mahkota bunga) dalam satuan cm.                 |
| **Species**       | VARCHAR(20) | Label kelas spesies bunga (*Setosa, Versicolor, Virginica*). |

Catatan:

*   Sepal adalah kelopak bunga yang melindungi kuncup.
*   Petal adalah mahkota bunga yang berwarna, biasanya paling bervariasi antar spesies.
*   Species adalah target label yang akan diprediksi.

## Kesimpulan Keseluruhan – Dataset Iris

Distribusi Data per Spesies
Dataset terdiri dari 150 data dengan jumlah seimbang untuk tiap spesies: Iris-setosa (50), Iris-versicolor (50), dan Iris-virginica (50). Hal ini menunjukkan tidak ada masalah class imbalance, sehingga dataset siap digunakan untuk analisis maupun pemodelan.

Petal Features (Kelopak) → Fitur Paling Penting

* talLengthCm memiliki rentang nilai yang lebar (±1 cm hingga hampir 7 cm), dengan rata-rata sekitar 3,7 cm. Variasi ini sangat jelas antar spesies, sehingga menjadi fitur utama untuk klasifikasi.

* PetalWidthCm juga memiliki perbedaan signifikan (±0,1 cm hingga 2,5 cm, rata-rata sekitar 1,2 cm). Sama seperti panjang kelopak, lebar kelopak adalah fitur kunci dalam membedakan spesies.

Sepal Features (Daun Mahkota) → Fitur Pendukung

* SepalLengthCm memiliki rentang 4,3 – 7,9 cm dengan rata-rata 5,8 cm. Meskipun ada variasi antar spesies, terdapat overlap nilai yang cukup besar, sehingga hanya cukup membantu sebagai fitur tambahan.

* SepalWidthCm memiliki rentang 2,0 – 4,4 cm dengan rata-rata sekitar 3,0 cm. Sama seperti panjang sepal, fitur ini kurang diskriminatif karena overlap antar spesies cukup tinggi.

Pola Utama

* Iris-setosa → ciri khas kelopak pendek dan sempit.

* Iris-versicolor → kelopak sedang.

* Iris-virginica → kelopak panjang dan lebar.
Pola ini menjelaskan mengapa fitur petal lebih efektif untuk membedakan spesies dibandingkan sepal.
